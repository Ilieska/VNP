```python
countries_of_interest = ['CAN', 'RUS']

# Step 2: Filter the dataset for these countries
filtered_df = data[data['Country Code'].isin(countries_of_interest)]

# Step 3: Remove duplicate URLs
filtered_df_unique = filtered_df.drop_duplicates(subset=['SOURCEURL'])

# Step 4: Ensure full URLs are displayed
pd.set_option('display.max_colwidth', None)

# Step 5: Print the filtered and deduplicated DataFrame
print("Source URLs for selected countries:")
print(filtered_df_unique[['Country Code', 'SOURCEURL']])
```

```python
url = filtered_df_unique['SOURCEURL']

def get_text_from_url(url):
    """Fetch and extract text content from a URL."""
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        paragraphs = soup.find_all('p')
        text = ' '.join(para.get_text() for para in paragraphs)
        return text
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return ""

def analyze_sentiment(text):
    """Perform sentiment analysis using VADER."""
    analyzer = SentimentIntensityAnalyzer()
    sentiment_score = analyzer.polarity_scores(text)
    return sentiment_score

# Analyze sentiment for each unique URL
sentiment_results = []
for _, row in filtered_df_unique.iterrows():
    url = row['SOURCEURL']
    country = row['Country Code']
    print(f"Processing URL: {url}")

    # Extract text from URL
    text = get_text_from_url(url)

    # Perform sentiment analysis
    sentiment = analyze_sentiment(text)

    # Store the results
    sentiment_results.append({
        'Country Code': country,
        'SOURCEURL': url,
        'Sentiment': sentiment
    })

# Create a DataFrame for sentiment results
sentiment_df = pd.DataFrame(sentiment_results)

# Print sentiment results
print("\nSentiment Analysis Results:")
print(sentiment_df)
```
